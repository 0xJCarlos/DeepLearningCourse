```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0,1,3]
y = [0,2,1]

plt.plot(x,y,'mo-') # m is magenta s are squares - is a line --dashed line : dotted line
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(np.random.randn(10))

plt.subplot(2,2,4)
plt.plot(np.random.randn(10))

plt.show()
```

```python
fig,ax = plt.subplots(1,2, figsize=(10,4))

x = np.arange(10)

ax[0].plot(x,x**2,'b')
ax[0].plot(x,x,'k')


ax[1].plot(x,np.sqrt(x),'r')

plt.show()
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))

plt.tight_layout()
plt.show()
```

```python
# Exercise create a 3,3 subplot geometry in a for loob

fig, ax = plt.subplots(3,3, figsize=(7,7))

for i in ax.flatten():
    i.plot(np.random.randn(3,9))

plt.show()
```
