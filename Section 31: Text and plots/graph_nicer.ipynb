```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x**2')
plt.plot(x,x**3,label='y=x**3')

plt.legend()
plt.xlabel('x')
plt.ylabel('y=f(x)')

plt.title("Awesome plot")

plt.xlim([x[0],x[-1]])
plt.ylim([-10,10])

#plt.axis('square')

plt.gca().set_aspect(.3)

plt.show()
```

```python
fig,ax = plt.subplots(1)

ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y=x**2')
ax.plot(x,x**3,label='y=x**3')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y=f(x)')

ax.set_title("Awesome plot")

ax.set_xlim([x[0],x[-1]])
ax.set_ylim([-10,10])

#plt.axis('square')

plt.gca().set_aspect(.3)

plt.show()
```

```python
# Exercise
fig,ax = plt.subplots()

x = np.linspace(0,10,100)

for i in np.linspace(0,1,50):
    ax.plot(x,x**i,color=[i/2,0,i])

plt.show()
```
