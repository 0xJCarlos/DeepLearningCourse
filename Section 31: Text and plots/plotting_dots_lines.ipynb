```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(3,4,'r^', label='red triangle')
plt.plot(2,4.2,'bo',label='blue circle')
plt.legend() #call to display legend
plt.show()

#figures o = circle s = square p = pentagram ^ = triangle
#color first letter before figure
```

```python
plt.plot(3,4,'r^')
plt.plot(2,4.2,'bo')
plt.legend(['red','blue']) #shows in order of plots
plt.show()

#figures o = circle s = square p = pentagram ^ = triangle
#color first letter before figure
```

```python
x = [0,1,3]
y = [0,2,1]

plt.plot(x,y,'mo-') # m is magenta s are squares - is a line --dashed line : dotted line
plt.show()
```

```python
import numpy as np 
x = np.linspace(0,3*np.pi,101)
y = np.sin(x)

plt.plot(x,y,'m:') # m is magenta s are squares - is a line --dashed line : dotted line
plt.show()
```
