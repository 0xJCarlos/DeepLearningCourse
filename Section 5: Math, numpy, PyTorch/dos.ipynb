```python
#imports
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Definir set de puntos a evaluar
x = np.linspace(.0001, 1, 200)

#Computar su logaritmo
logx = np.log(x)

#Plot!
fig = plt.figure(figsize=(10,4))

plt.rcParams.update({'font.size':15}) #Tamaño de fuente

plt.plot(x, logx, 'ks-', markerfacecolor='w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
#Demostrar que log y exp son inversos

#redefinir con menos puntos

x = np.linspace(.0001, 1 , 20)

#log y exp
logx = np.log(x)
expx = np.exp(x)

#plot

plt.plot(x,x,color=[.8,.8,.8])
plt.plot(x,np.exp(logx),'o', markersize=8)
plt.plot(x,np.log(expx),'x', markersize=8)
plt.xlabel('x')
plt.ylabel('f(g(x))')
plt.legend(['unity','exp(log(x))','log(exp(ex))'])
plt.show()
```

```python
#Prboabilidad de que algo pase
p = .25

#formula INCORRECTA
H = -(p*np.log(p) )
print('Entropia equivocada: ' + str(H))
```

```python
#Calcular entropia de la forma correcta
x = [.25, .75]

H = 0
for p in x:
    H -= p*np.log(p)

print('Entropia correcta: ' + str(H))
```

```python
#También correcta, para eventos N=2, es decir, 1 evento y la posibilidad que suceda Y que NO suceda

#Binary entropy
H = -( p*np.log(p) + (1-p)*np.log(1-p) )
print('Correct entropy: ' + str(H))
```

```python
#CROSS-ENTROPY
#nota: Todas las probabilidades deben sumar a 1 

p = [1,0] # suma a 1
q = [.25,.75] # suma a 1

H = 0
for i in range(len(p)):
    H -= p[i]*np.log(q[i])

print('Cross entropy: ' + str(H))
```

```python
 #También correcta para eventos N=2
 H = -( p[0]*np.log(q[0]+p[1]*np.log(q[1])))
 print('Correct entropy: ' + str(H))

 #Simplificacion
 H = -np.log(q[0])
 print ('Simplificación manual: ' + str(H))
```
