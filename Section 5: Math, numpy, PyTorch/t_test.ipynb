```python
#imports
import numpy as np 
import matplotlib.pyplot as plt

import scipy.stats as stats 
```

```python
# Parameters
n1 = 20 #Samples in dataset 1 
n2 = 60 # ... and 2
mu1 = 6  #Population mean in dataset 1
mu2 = 1.02  #Population mean in dataset 2

#Generate the data

data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

#Plot them

plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n1), data1, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
```

## T-Test time!

```python
#_ind = independent samples
t, p = stats.ttest_ind(data1, data2)
print(t)
print(p)
```

```python
#common way to t-test in a plot
fig = plt.figure(figsize=(5,5))
plt.rcParams.update({'font.size':12})

plt.plot(0+np.random.randn(n1)/15, data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1+np.random.randn(n2)/12, data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1', 'Group 2'])

#Set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p={p:.3f}')

plt.show()
```
